```python
medals = medals.merge(drinks[['country','continent']],how='left',left_on='Country',right_on='country')
```

```python
medals = medals.drop('country', axis=1)
```

```python
medals
```

```python
medals.loc[medals.continent.isnull() , 'Country']
```

# 8

```python
medals = medals.set_index('Country')
medals.loc["Australasia",'continent'] = 'OC'
medals.loc[["Canada","United States","Mexico"],'continent'] = 'NA'
medals.loc[["Cuba","Jamaica"],'continent'] = 'SA'
medals.loc[["Czechoslovakia","Yugoslavia","United Team of Germany","East Germany","West Germany","Great Britain","Russia",
            "Russian Empire","Soviet Union"],'continent'] = 'EU'
medals.loc[medals.continent.isna() == True , 'continent'] = 'OT'
medals = medals.reset_index('Country')
wide_table = medals
medals
```

# 9

```python
medals = medals.set_index(['continent' , 'Country'])
```

```python
medals
```

# 10

```python
medals = medals.sort_index(level=[0,1], ascending=[False,True])
medals 
```

# 11

```python
medals = medals.drop('Code',axis = 1)
columns = pd.MultiIndex.from_tuples([('Summer','Game'),('Summer','Gold'),('Summer','Silver'),('Summer','Bronze')
                                    ,('Winter','Game'),('Winter','Gold'),('Winter','Silver'),('Winter','Bronze')])
medals.columns = columns
medals
```

# 12

```python
medals = medals.swaplevel(0,1,axis=1)
medals
```

# 13

```python
long_table = medals.stack().reset_index().melt(id_vars=['continent','Country','level_2','Game'])
long_table.columns=['continent','Country','Season','Game','Medal','Count']
long_table
```
